```python
import numpy as np
```

```python
import random
```

```python
random.seed(1001)
```

```python
print(random.random())
```

```python
x = np.random.normal(loc= 300.0, size=1000)
x
```

```python
np.mean(x)
```

```python
np.var(x)
```

```python
print(np.mean(x))
```

```python
sample_mean = []
sample_mean
```

```python
for i in range(50):
  y = random.sample(x.tolist(), 4)
  avg = np.mean(y)
  sample_mean.append(avg)
```

```python
sample_mean
```

```python
se_mean = np.sqrt(np.var(sample_mean))
print(se_mean)
```

```python
np.mean(sample_mean)
```

```python
sample_t = np.mean(sample_mean)/se_mean
sample_t
```
